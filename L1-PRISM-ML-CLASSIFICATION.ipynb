```python
from sklearn.linear_model import LinearRegression
import pandas as pd 
import seaborn as sns
from sklearn import preprocessing
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np 
```

```python
iris=pd.read_csv("Iris(1).csv")
```

```python
iris.head(10)
```

```python
iris.drop("Id",axis=1,inplace=True);
```

```python
iris.head(10)
```

```python
target=iris["Species"]
target.shape
```

```python
data=iris.drop(["Species"],axis=1)
data.shape
```

```python
data.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data=scaler.fit_transform(data)
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
tr=le.fit_transform(target)
tr
```

```python
data_train,data_test,target_train,target_test=train_test_split(data,target,test_size=0.3,random_state=42)
print(f"Data Training is -- {data_train.shape}")
print(f"Data Testing is -- {data_test.shape}")
print(f"Target Training is -- {target_train.shape}")
print(f"Target Testing is -- {target_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(data_train,target_train)
```

```python
target_pred=model.predict(data_test)
```

```python
target_pred
```

```python
import seaborn as sns 
import matplotlib.pyplot as pyplot
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(classification_report(target_test,target_pred))
```

```python
cm=confusion_matrix(target_test,target_pred)
ax=plt.subplot()
sns.heatmap(cm,annot=True,ax=ax,cmap="Blues",fmt="g")
ax.set_xlabel("Predicited Labels")
ax.set_ylabel("True Labels")
ax.set_title("Confusion Matrix")


ax.xaxis.set_ticklabels(["Setosa","Versicolour","Virginica"])
ax.yaxis.set_ticklabels(["Setosa","Versicolour","Virginica"])
plt.show()
```
